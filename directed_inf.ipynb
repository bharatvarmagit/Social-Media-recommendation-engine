```python
from collections import defaultdict
v_list=[]
adj_list =defaultdict(list)
file=open("Wiki-Vote.txt","r")
lines=file.readlines()
for line in lines:
    word=line.split()
    if word[1] not in adj_list[word[0]]:
        adj_list[word[0]].append(word[1])
    if word[0] not in v_list:
        v_list.append(word[0])
    if word[1] not in v_list:
        v_list.append(word[1])
file.close()
print(len(adj_list))
print(len(v_list))
```

```python
import operator 
from collections import defaultdict
def counter(adj_list,v):
    count=0
    for key in adj_list.keys():
        if v in adj_list[key]:
            count+=1
    return count
        
influencers=defaultdict(list)
for key in adj_list.keys():
    influencers[key]=counter(adj_list,key)
inf=dict(sorted(influencers.items(),key=operator.itemgetter(1),reverse=True)[:100])
inf
```
