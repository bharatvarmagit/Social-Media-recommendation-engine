```python
from collections import defaultdict
v_list=[]
adj_list =defaultdict(list)
file=open("Wiki-Vote.txt","r")
lines=file.readlines()
for line in lines:
    word=line.split()
    if word[1] not in adj_list[word[0]]:
        adj_list[word[0]].append(word[1])
    if word[0] not in v_list:
        v_list.append(word[0])
    if word[1] not in v_list:
        v_list.append(word[1])
file.close()
#print(len(adj_list))
#print(len(v_list))



#find connected components
def scc(a_list): 
    scc=[]
    stack = [] 
    # Mark all the vertices as not visited (For first DFS) 
    visited =set() 
    # Fill vertices in stack according to their finishing 
    # times 
    for i in sorted(v_list): 
        if i not in visited: 
            dfs1(a_list,i, visited, stack) 
  
    # Create a reversed graph 
    gr = getTranspose(a_list) 
    
           
         # Mark all the vertices as not visited (For second DFS) 
    visit =set()
  
         # Now process all vertices in order defined by Stack 
    while stack: 
        i = stack.pop() 
        
        if i not in visit: 
            temp=[]
            
            
            scc.append(dfs2(gr, i, visit,temp)) 
    return scc
            
        
def dfs1(a_list,v,visited, stack): 
    # Mark the current node as visited  
    visited.add(v)

    adj_f=[i for i in a_list[v]]
    #Recur for all the vertices adjacent to this vertex 
    for i in adj_f: 
        if i not in visited: 
            dfs1(a_list, i, visited, stack) 
            #add vertex to stack if no more new vertices can be traverserd through v
    stack = stack.append(v) 
    
def dfs2(gr,v,visit,temp): 
    # Mark the current node as visited and print it 
    visit.add(v)
   
    
    adj_v=[i for i in gr[str(v)]]
    temp.append(v)
   
    
    
    for i in adj_v: 
        
        if i not in visit:
           
            temp=dfs2(gr,i,visit,temp)
    return temp       

def getTranspose(a_list): 
    adj_list_t =defaultdict(list) 
  
    
    for key in list(a_list): 
        adj_v=[k for k in a_list[key]]
        for j in adj_v:
            adj_list_t[j].append(key)
             
    return adj_list_t
        
scc_list=scc(adj_list)
dictscc=defaultdict(list)
for i,v in enumerate(scc_list):
    dictscc[i+1] = [i for i in scc_list[i]]
dictscc
```

```python
from collections import defaultdict
vertices=list()
adj_list =defaultdict(list)
file=open("walk.txt","r")
lines=file.readlines()
for line in lines:
    word=line.split()
    if word[1] not in adj_list[word[0]]:
        adj_list[word[0]].append(word[1])
    if word[0] not in vertices:
        vertices.append(word[0])
    if word[1] not in vertices:
        vertices.append(word[1])
file.close()
vertices= list(map(int, vertices)) 

def walk(u,v,numwalks,adj_list,vertices):
    vs=max(vertices)+1
    mat= [[[0 for k in range(numwalks+1)] for j in range(vs)] for i in range(vs)]
    for k in range(numwalks+1):
        
          for i in range(vs):
                
                for j in range(vs):
                    
                    if k == 0 and i == j:
                        
                        mat[i][j][k]=1
                        
                    elif k==1 and str(i) in adj_list and str(j) in adj_list[str(i)]:
                        mat[i][j][k]=1
                    
                    elif k>1:
                        for a in range(vs):
                            if str(i) in adj_list and str(a) in adj_list[str(i)]:
                                mat[i][j][k] += mat[a][j][k-1]
    return mat[u][v][numwalks]
print(walk(0,3,2,adj_list,vertices))
```
