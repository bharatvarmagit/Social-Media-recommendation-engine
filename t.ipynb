```python
from collections import defaultdict
vertices=list()
adj_list =defaultdict(list)
file=open("direct.txt","r")
lines=file.readlines()
for line in lines:
    word=line.split()
    if word[1] not in adj_list[word[0]]:
        adj_list[word[0]].append(word[1])
    if word[0] not in vertices:
        vertices.append(word[0])
    if word[1] not in vertices:
        vertices.append(word[1])
file.close()
vertices= list(map(int, vertices)) 

def countwalks(verties,adj_list, u, v, k): 
    vs=max(vertices)+1
  
    # Base cases 
    if (k == 0 and u == v): 
        return 1
    if k == 1 and str(u) in adj_list and str(v) in adj_list[str(u)]: 
        return 1
    if (k <= 0): 
        return 0
      
    # Initialize result 
    count = 0
      
    # Go to all adjacents of u and recur 
    for i in range(vs): 
          
        # Check if is adjacent of u 
        if str(u) in adj_list and str(i) in adj_list[str(u)]: 
            count += countwalks(vertices,adj_list, i, v, k-1) 
      
    return count 
print(countwalks(vertices,adj_list,0,3,2))



                  


```
