```python
from collections import defaultdict
v_list=[]
adj_list =defaultdict(list)
file=open("direct.txt","r")
lines=file.readlines()
for line in lines:
    word=line.split()
    if word[1] not in adj_list[word[0]]:
        adj_list[word[0]].append(word[1])
    if word[0] not in v_list:
        v_list.append(word[0])
    if word[1] not in v_list:
        v_list.append(word[1])
file.close()
#print(len(adj_list))
#print(len(v_list))



#find connected components
def printSCCs(a_list): 
    print("vlist =")
    print(v_list)
    stack = [] 
    # Mark all the vertices as not visited (For first DFS) 
    visited =set() 
    # Fill vertices in stack according to their finishing 
    # times 
    for i in sorted(v_list): 
        if i not in visited: 
            fillOrder(a_list,i, visited, stack) 
  
    # Create a reversed graph 
    gr = getTranspose(a_list) 
    print(f'gr = {gr}')
   
    print("here comes the stack")
    print(stack)
           
         # Mark all the vertices as not visited (For second DFS) 
    visit =set()
  
         # Now process all vertices in order defined by Stack 
    while stack: 
        i = stack.pop() 
        
        if i not in visit: 
            
            DFSUtil(gr, i, visit) 
            print("next component")
        
def fillOrder(a_list,v,visited, stack): 
    # Mark the current node as visited  
    visited.add(v)

    adj_f=[i for i in a_list[v]]
    #Recur for all the vertices adjacent to this vertex 
    for i in adj_f: 
        if i not in visited: 
            fillOrder(a_list, i, visited, stack) 
            
    stack = stack.append(v) 
    
def DFSUtil(gr,v,visit): 
    # Mark the current node as visited and print it 
    visit.add(v)
   
    
    adj_v=[i for i in gr[str(v)]]
    print(v)
   
    
    #Recur for all the vertices adjacent to this vertex 
    for i in adj_v: 
        
        if i not in visit:
           
            DFSUtil(gr,i,visit)
            

def getTranspose(a_list): 
    adj_list_t =defaultdict(list) 
  
    # Recur for all the vertices adjacent to this vertex 
    for key in list(a_list): 
        adj_v=[k for k in a_list[key]]
        for j in adj_v:
            adj_list_t[j].append(key)
             
    return adj_list_t
        
printSCCs(adj_list)
```
